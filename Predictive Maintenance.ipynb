```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Overlook _ Training dataset

```python
df = pd.read_csv('train_FD004.txt', sep= ' ', header = None)
```

```python
df.head()
```

## Naming the columns of the Dataset

```python
df = df[[f for f in range(0,26)]]  #removing the last two redundant column
df.columns = ['ID', 'Cycle', 'openset1', 'openset2', 'openset3', 'sensormesurment1','sensormesurment2','sensormesurment3','sensormesurment4','sensormesurment5','sensormesurment6','sensormesurment7','sensormesurment8','sensormesurment9','sensormesurment10','sensormesurment11','sensormesurment12','sensormesurment13','sensormesurment14','sensormesurment15','sensormesurment16','sensormesurment17','sensormesurment18','sensormesurment19','sensormesurment20','sensormesurment21']
df
```

```python
df[df['ID'] == 1]
```

```python
EOL =[]  # Adding new column of End Of Life
for ii in df['ID']:
    EOL.append( ((df[df['ID'] == ii]["Cycle"]).values)[-1])
```

```python
len(EOL)
```

```python
df['EOL'] = EOL
```

```python
df
```

## Classification of engine stage depend on the remaining life cycle ratio

### There are threee stage in which engine is classified:
#### Safe LCR <=0.6, Moderate LCR <=0.8, High risk LCR >0.8

```python
df['LCR'] = df['Cycle'].div(df['EOL'])
df = df.drop(columns = ['ID','EOL'])
df
```

```python
labels = []
for i in range (0,len(df)):
    if np.array(df['LCR'])[i] <= 0.6:
        labels.append(0)
    elif np.array(df['LCR'])[i] <= 0.8:
        labels.append(1)
    else:
        labels.append(2)

df['labels'] = labels
df = df.drop(columns = ["LCR"])
df
```

```python
#df.to_csv('Traning_data4.csv', index = None) 
```

### Importing all the training data after cleaning(preprocessing)

```python
df1 = pd.read_csv('Traning_data1.csv')
df2 = pd.read_csv('Traning_data2.csv')
df3 = pd.read_csv('Traning_data3.csv')
df4 = pd.read_csv('Traning_data4.csv')
    
```

```python
df = pd.concat([df1,df2,df3,df4])
```

```python
df
```

#### Dependent variable and independent variable

```python
X = df.iloc[:,0:-1]
Y = df.iloc[:,-1]
X
```

```python
Y
```

### Training of the data and test by Random Forest algorithm

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.1, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(X_train, Y_train)
```

```python
y_pred_RF = RF.predict(X_test)

cm = pd.DataFrame( confusion_matrix(Y_test, y_pred_RF, normalize = 'true')*100)
colormap = sns.color_palette('Reds')
sns.heatmap(cm, annot = True, cmap = colormap)

accuracy_score(Y_test, y_pred_RF)

```

```python
y_pred_RF = RF.predict(X_train)

cm = pd.DataFrame( confusion_matrix(Y_train, y_pred_RF, normalize = 'true')*100)
colormap = sns.color_palette('Blues')
sns.heatmap(cm, annot = True, cmap = colormap)

accuracy_score(Y_train, y_pred_RF)
```
